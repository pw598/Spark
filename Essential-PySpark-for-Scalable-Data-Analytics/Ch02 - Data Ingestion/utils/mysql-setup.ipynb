###Utility notebook to install MySQL Server
This notebook:
1. Installs a `mysql-server` locally on your `driver` node.
2. Creates a database called `pysparkdb`.
3. Creates a table called `authors` and inserts some sample data.

```
%sh
sudo debconf-set-selections <<< 'mysql-server mysql-server/root_password password root'
sudo debconf-set-selections <<< 'mysql-server mysql-server/root_password_again password root'
sudo apt-get -y install mysql-server
service mysql restart
```

```
%sh service mysql status
```

```
%sh 
mysql -u root -proot -e "create database pysparkdb;"
mysql -u root -proot -e """drop table if exists pysparkdb.authors; 
CREATE TABLE pysparkdb.authors (
  uid INT NOT NULL AUTO_INCREMENT,
  last_name VARCHAR( 50 ),
  first_name VARCHAR( 50 ),
  birth_month VARCHAR( 50 ),
  PRIMARY KEY (uid)
)"""
```

```
%sh mysql -u root -proot -e """
INSERT INTO pysparkdb.authors (last_name, first_name, birth_month)
VALUES
  ('Asimov', 'Issac', 'January'),
  ('Brown', 'Dan', 'June'),
  ('Baldacci', 'David', 'August'),
  ('Burroughs', 'Edgar Rice', 'September'),
  ('Clarke', 'Arthur C.', 'December'),
  ('Cook', 'Robin', 'May'),
  ('Dickens', 'Charles', 'February'),
  ('Forster', 'E.M.', 'January'),
  ('Lahiri', 'Jhumpa', 'July'),
  ('Rand', 'Ayn', 'February'),
  ('Salinger', 'J.D.', 'January'),
  ('Seuss', 'Dr.', 'March'),
  ('Stoker', 'Bram', 'November'),
  ('Wilde', 'Oscar', 'October'),
  ('Wordsworth', 'William', 'April');
"""
```
